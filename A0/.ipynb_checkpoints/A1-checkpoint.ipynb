```python
from numpy import genfromtxt
import matplotlib.pyplot as plt
import numpy as np
import math
import numpy.ma as ma
from sklearn.decomposition import PCA
import umap
from sklearn.manifold import TSNE

N = 10
def main():
    train_in, test_in, train_out, test_out = load_data()
    print(train_in.shape)
    center = center_compute(train_in, train_out)
    matrix = dis_between_centers(center)
    
    '''mnist = fetch_openml('mnist_784', version=1, cache=True)
    x = np.array(mnist.data)
    y = np.array(mnist.target).astype(int)
    m, n = x.shape
    
    n = 10
    img = np.zeros((30 * n, 30 * n))
    
    for i in range(n):
        ix = 30 * i + 1
        for j in range(n):
            iy = 30 * j + 1
            img[ix : ix + 28, iy : iy + 28] = x[i * n + j].reshape(28, 28)
    
    plt.imshow(img, cmap=plt.cm.binary)
    plt.set_title(f"{y}",fontsize=10)
    plt.axis("off")
    plt.show()
    scaler = StandardScaler()
    x_s = scaler.fit_transform(x)'''
    
    # pca visualize
    m, n = train_in.shape
    pca = PCA(n_components=2)
    pca.fit(train_in)
    x_p = pca.transform(train_in)
    r1, a1 = digit_recognization(center, x_p, train_out)
    
    # randomly visualize
    fig, axes = plt.subplots(8,8, figsize=(5,5))
    fig.tight_layout(pad=0.13,rect=[0, 0.03, 1, 0.91]) #[left, bottom, right, top]
    for i, ax in enumerate(axes.flat):
        random_index = np.random.randint(m)
        # reshape
        x_random_reshaped = train_in[random_index].reshape((28,28))
        ax.imshow(x_random_reshaped, cmap='gray')
        ax.set_title(f"{train_out[random_index]},{r1[random_index]}",fontsize=10)
        ax.set_axis_off()
    fig.suptitle("Label, recognization", fontsize=14)
    plt.show()
    
    # U-map
    u = umap.UMAP(n_components=2)
    x_u = u.fit_transform(train_in)    
    r2, a2 = digit_recognization(center, x_u, train_out)
    
    # TSNE
    x_t = TSNE(n_components=2).fit_transform(train_in)
    r3, a3 = digit_recognization(center, x_t, train_out)

def load_data():
    train_in = genfromtxt('./data/train_in.csv', delimiter=',')
    test_in = genfromtxt('./data/test_in.csv', delimiter=',')
    train_out = genfromtxt('./data/train_out.csv')
    test_out = genfromtxt('./data/test_out.csv')
    return train_in, test_in, train_out, test_out

def center_compute(train_in, train_out):
    number = dict.fromkeys(train_out[:].astype(int))
    center = np.zeros(shape=(N, train_in.shape[1]))
    for i in range(N):
        l = []
        for j, character in enumerate(train_out):
            if character == i:
                l.append(j)
        number[i] = l
    for d in range(N):
        center[d] = np.mean(train_in[number[d]], axis=0)
    return center

def digit_recognization(center, test_in, test_out):
    recognization = np.zeros_like(test_out)
    for i, t in enumerate(test_in):
        distance = []
        for j, c in enumerate(center):
            distance.append(math.dist(t, c))
        recognization[i] = np.argmin(np.array(distance))
    accuracy = np.sum(recognization != test_out)/len(test_out)
    return recognization, accuracy

def dis_between_centers(center):
    dis_matrix = np.zeros(shape=(N, N))
    for i, ci in enumerate(center):
        for j, cj in enumerate(center):
            dis_matrix[i, j] = math.dist(ci, cj)
    
    m_matrix = ma.masked_array(dis_matrix, mask=np.identity(dis_matrix.shape[0]))
    indice = np.where(m_matrix == np.min(m_matrix))
    return dis_matrix, indice

main()
```
